```python
#9
import matplotlib.pyplot as plt

# The slices will be ordered and plotted counter-clockwise.
labels='bye runs','wide runs','legbye runs','noball runs','penalty runs'
sizee=[680,5161,3056,612,10]
colors = ['yellowgreen', 'gold', 'lightskyblue', 'lightcoral','red']
explode = (0, 0, 0, 0,0)  # explode a slice if required
fig1, ax1 = plt.subplots()

ax1.pie(sizee, explode=explode, labels=labels, colors=colors,autopct='%1.1f%%', shadow=True)
        
#draw a circle at the center of pie to make it look like a donut
centre_circle = plt.Circle((0,0),0.75,color='black', fc='white',linewidth=1.25)
fig = plt.gcf()
fig.gca().add_artist(centre_circle)


# Set aspect ratio to be equal so that pie is drawn as a circle.
plt.axis('equal')
plt.show()  
```
